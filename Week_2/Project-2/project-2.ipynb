```python
import numpy as np
username = input("Enter your username: ")

def login():

    image_path = f"C:/Users/Oluwanifemi Oyewole/Desktop/PAU/YEAR 4/Semester 2/CSC 418 - Special Topics in Computer Science/Practice/noyewoleCSC418/Week_2/Project-2/img/{username}.jpg"

    if os.path.exists(image_path):
        password = input(
            "Enter your password (number of alphabets in your username): ")

        if len(username) == int(password):
            print(f"Welcome, {username}!")
            print("")
            enhancement = input("Select the enhancement method you want to use: \n\nVisualize Color Space (Press 1) ||| \nArithmetic Operation on Images (Press 2) ||| \nAdjusting Brightness and Contrast (Press 3) ||| \nSharpening Images (Press 4) ||| \nRemoving Noise (Press 5) ||| \nImage Scaling (Press 6) ||| \nInverse Transform (Press 7)")

            if int(enhancement) == 1:
                visualizeColorSpaces()
            if int(enhancement) == 2:
                arithmeticOperationOnImages()
            if int(enhancement) == 3:
                adjBrightnessAndContrast()
            if int(enhancement) == 4:
                sharpenImage()
            if int(enhancement) == 5:
                removeNoise()
            if int(enhancement) == 6:
                imageScaling()
            if int(enhancement) == 7:
                inverseTransform()
            # else:
            #     print("Invalid entry.")

        else:
            print("Incorrect password.")
    else:
        print("User not found.")


def visualizeColorSpaces():
    image = cv2.imread(f'img/{username}.jpg')
    # the split method separates color spaces
    B, G, R = cv2.split(image)

    # corresponding channels are separated
#     cv2.imshow("original", image)
#     cv2.waitKey(0)
#     cv2.imshow("blue", B)
#     cv2.waitKey(0)
#     cv2.imshow("green", G)
#     cv2.waitKey(0)
#     cv2.imshow("red", R)
#     cv2.waitKey(0)
#     cv2.destroyAllWindows()

    # plotting
    plt.subplot(2, 2, 1)
    plt.title(f"{username}")
    plt.imshow(image)

    plt.subplot(2, 2, 2)
    plt.title("Blue")
    plt.imshow(B)

    plt.subplot(2, 2, 3)
    plt.title("Red")
    plt.imshow(R)

    plt.subplot(2, 2, 4)
    plt.title("Green")
    plt.imshow(G)

def arithmeticOperationOnImages():
    # path to input images
    image1 = cv2.imread(f'img/{username}.jpg')
    image2 = cv2.imread('img/shrek2.jpg')
    # resize images
    image1 = cv2.resize(image1, (500, 400))
    image2 = cv2.resize(image2, (500, 400))
    # cv2.addWeighted with added parameters
    addImage = cv2.addWeighted(image1, 0.5, image2, 0.6, 0)
    # window showing output image
    cv2.imshow('Weighted Image', addImage)
    # de-allocate any associated memory usage
    if cv2.waitKey(0) & 0xff == 27:
        cv2.destroyAllWindows()

def adjBrightnessAndContrast():
    import matplotlib.pyplot as plt
    # load the image
    image = cv2.imread(f'img/{username}.jpg')
    # plot the original image
    plt.subplot(1, 2, 1)
    plt.title("Original")
    plt.imshow(image)
    # adjust the brightness and contrast
    brightness = 5
    contrast = 1.5
    image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness)
    # save the image
    cv2.imwrite('img/saved-imgs-after-running/contrast_img.jpg', image2)
    # plot the contrast image
    plt.subplot(1, 2, 2)
    plt.title("Brightness & Contrast")
    plt.imshow(image2)
    plt.show

def sharpenImage():
    # load the image
    image = cv2.imread(f'img/{username}.jpg')
    # plot the original image
    plt.subplot(1,2,1)
    plt.title("Original")
    plt.imshow(image)
    # create the sharpening kernel
    kernel = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
    # sharpening image
    sharpened_image = cv2.filter2D(image, -1, kernel)
    # save the image
    cv2.imwrite('img/saved-imgs-after-running/sharpened_image.jpg', sharpened_image)
    # plot the sharpened image
    plt.subplot(1,2,2)
    plt.title("Sharpening")
    plt.imshow(sharpened_image)
    plt.show()

def removeNoise():
    # load the image
    image = cv2.imread(f"img/{username}.jpg")
    # plot the original image
    plt.subplot(1,2,1)
    plt.title("Original")
    plt.imshow(image)
    # remove noise using a median filter
    filtered_image = cv2.medianBlur(image, 15)
    # save the image
    cv2.imwrite('img/saved-imgs-after-running/Median-Blur.jpg', filtered_image)
    # plot blurred image
    plt.subplot(1,2,2)
    plt.title("Median Blur")
    plt.imshow(filtered_image)
    plt.show()

def imageScaling():
    # load the image
    image = cv2.imread(f'img/{username}.jpg')
    # plot the original image
    plt.subplot(1,2,1)
    plt.title("Original")
    plt.imshow(image)
    # scale image by factor of two along both axes
    scaled_image = cv2.resize(image, None, fx=2, fy=2)
    # save the image
    cv2.imwrite('img/saved-imgs-after-running/Scaled.jpg', scaled_image)
    # plot scaled image
    plt.subplot(1,2,2)
    plt.title("Scaled")
    plt.imshow(scaled_image)
    plt.show()

def inverseTransform():
    # load the image
    image = cv2.imread(f'img/{username}.jpg')
    # plot the original image
    plt.subplot(1,2,1)
    plt.title("Original")
    plt.imshow(image)
    # inverse by subtracting from 255
    inverse_image = 255 - image
    # save the image
    cv2.imwrite('img/saved-imgs-after-running/inverse_image.jpg', inverse_image)
    # plot inverted image
    plt.subplot(1,2,2)
    plt.title("Inverse color")
    plt.imshow(inverse_image)
    plt.show()

login()
```
