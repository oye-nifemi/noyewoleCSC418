## Image Scaling

```python
import cv2
import matplotlib.pyplot as plt
# load the image
image = cv2.imread('img/sst.jpg')
# plot the original image
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(image)
# scale image by factor of two along both axes
scaled_image = cv2.resize(image, None, fx=2, fy=2)
# save the image
cv2.imwrite('img/Scaled.jpg', scaled_image)
# plot scaled image
plt.subplot(1,2,2)
plt.title("Scaled")
plt.imshow(scaled_image)
plt.show()
```
